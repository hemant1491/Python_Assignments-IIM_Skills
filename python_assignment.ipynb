Q : Write a Python script that prints the numbers from 1 to 100, but for multiples of three print "Fizz" instead of the number and for the multiples of five print "Buzz". For numbers which are multiples of both three and five, print "FizzBuzz".

```python
for i in range(1,101):
    if i % 3 == 0 and i % 5 == 0:
        print("FizzBuzz")
    elif i % 5 == 0:
        print("Buzz")
    elif i % 3 == 0:
        print("Fizz")
    else:
        print(i)
```

Q : Write a Python script that calculates the factorial of a given number using a loop. Prompt the user to enter a number and display its factorial.

```python
# num = int(input("Enter your number"))
num = 6
facto = 1

while (num > 0):
    facto = facto * num
    num = num - 1
    
f"The factorial is {facto}"
```

Q : Write a Python script that takes a string input from the user and prints whether the string is a palindrome. Use conditional statements and loops in your solution.

```python
text = "racecar"

text_list = []

for i in text:
    text_list.append(i)
    
palindrome_text = "".join(reversed(text_list))

if text == palindrome_text:
    print("This is a palindrome text")
else:
    print("This isn't a palindrome text")
    
```

Q : Define a function in Python that takes a list of numbers and returns the sum of all even numbers in the list. Use a lambda function to filter out the even numbers before summing them.

```python
my_range = range(1,21)
my_list = list(my_range)

even_num = list(filter(lambda x : x % 2 == 0,my_list))
print(even_num)

sum = 0

for i in even_num:
    sum = sum + i
        
f"sum of even numbers is {sum}"
```

Q : Import the Titanic dataset (using pandas) into a Pandas DataFrame. Perform basic operations such as displaying the first few rows, checking for null values, and summarizing the DataFrameâ€™s statistics.

```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

titanic = pd.DataFrame(sns.load_dataset("titanic"))
titanic.head(10)
```

```python
titanic.info()
```

```python
titanic.describe()
```

```python
# Checking for null values
titanic.isnull().sum()
```

```python
# Visualisation for outliers if present

plt.figure(figsize=(10,7))
sns.boxplot(x= "age", data=titanic,)
plt.xlabel("Age")
plt.title("Box Plot for age")
plt.show()
```

```python
# Filling missing value in column age

def outliers(data):
    q1 = data.quantile(0.25)
    q3 = data.quantile(0.75)
    iqr = q3 - q1
    ub = q3 + (iqr * 1.5)
    lb = q1 - (iqr * 1.5)
    outliers = data[(data < lb) | (data > ub)]
    return outliers

age_outliers = outliers(titanic["age"])

def filling_missing_value(df, column, outliers):
    if not outliers.empty:
        df.fillna({column:df[column].median()}, inplace=True)
    else:
        df.fillna({column:df[column].mean()}, inplace=True)

filling_missing_value(titanic, "age", age_outliers)

titanic.head(10)
```

Q : Load the Iris dataset (using pandas) into a Pandas DataFrame and perform data cleaning operations such as removing duplicates, handling missing values, and converting data types. Provide code and explanations for each step.

```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Loading dataset
iris = pd.DataFrame(sns.load_dataset("iris"))
iris.head(10)
```

```python
# Checking for information
iris.info()
```

```python
# Describing the dataset
iris.describe()
```

```python
# Removing duplicates values in iris dataset from column species

iris.drop_duplicates(inplace=True)
iris
```

```python
# Defining the format for subplots
fig , axes = plt.subplots(2, 2, figsize=(15,10))

# Common title for all subplots
fig.suptitle("Box Plot for Sepal(Length & Width) and Petal(Length & Width)", fontsize=15)

sns.boxplot(data=iris, x="sepal_length", color="red", ax=axes[0,0])
sns.boxplot(data=iris, x="sepal_width", color="orange", ax=axes[0,1])
sns.boxplot(data=iris, x="petal_length", color="gold", ax=axes[1,0])
sns.boxplot(data=iris, x="petal_width", color="green", ax=axes[1,1])

plt.show()
```

```python
# Handling Missing value

def outliers(data):
    q1 = data.quantile(0.25)
    q3 = data.quantile(0.75)
    iqr = q3 - q1
    ub = q3 + (iqr * 1.5)
    lb = q1 - (iqr * 1.5)
    outliers = data[(data < lb) | (data > ub)]
    return outliers

tsl = outliers(iris["sepal_length"])
tsw = outliers(iris["sepal_width"])
tpl = outliers(iris["petal_length"])
tpw = outliers(iris["petal_width"])

def fill_missing_value(df, column, outliers):
    if not outliers.empty:
        df.fillna({column:df[column].median()}, inplace=True)
    else:
        df.fillna({column:df[column].mean()}, inplace=True)

fill_missing_value(iris, "sepal_length", tsl)
fill_missing_value(iris, "sepal_width", tsw)
fill_missing_value(iris, "petal_length", tpl)
fill_missing_value(iris, "petal_width",tpw)
```

```python
# Converting species from object type to string type

iris["species"] = iris["species"].convert_dtypes(str)
iris
```

Q : Implement a linear regression model using Scikit-Learn on the Boston Housing dataset (using sklearn). Use the dataset to predict the median value of owner-occupied homes (MEDV). Split the data into training and testing sets, train the model, and evaluate its performance using RMSE and MSE metrics. Provide a brief interpretation of the results.

```python
%matplotlib inline
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error, r2_score

# importing data 

url = "https://raw.githubusercontent.com/selva86/datasets/refs/heads/master/BostonHousing.csv"

bhd = pd.read_csv(url)
bhd.head(10)
```

```python
# information of the dataset
bhd.info()
```

```python
# Describing the dataset
bhd.describe()
```

```python
# Creating the histogram to know the distribution

plt.figure(figsize=(10,7))
sns.histplot(data=bhd, x="medv", bins=30, color="green", kde=True)
plt.show()
```

```python
# Creating Heat map to find out corelation between all neumerical variable
plt.figure(figsize=(12,10))
sns.heatmap(bhd.corr(), annot=True, cmap="bwr", fmt=".2f", linewidths=0.5)
plt.show()
```

medv has strong positive co-relation with "rm" and strong negative co-relation with "lstat".

```python
# Plotting the scatter plot to find out corelation between rm and lstat 

plt.figure(figsize=(10,7))
sns.scatterplot(data=bhd, x="rm", y="lstat", color="blue")
plt.title("Corelation between 'rm' and 'lstat'")
plt.show()
```

```python
X = bhd[["rm","lstat"]] # Independent variable
Y = bhd["medv"] # Dependent or target variable
X.head()
```

```python
Y.head()
```

```python
# Split the dataset in training and testing sets
from sklearn.model_selection import train_test_split

# test_size = 0.2 means training 80% and for testing 20%
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.2, random_state=42)
```

```python
# Train Model

# Initialize and train the model
model = LinearRegression()
model.fit(X_train, Y_train)

# Print the regression equation Y = mX + C

intercept = model.intercept_
coefficients = model.coef_

f"Regression equation: medv = {intercept:.2f} + {coefficients[0]:.2f} * rm + {coefficients[1]:.2f} * lstat"
```

```python
# Make prediction
# Predicting on training data is not advisable

Y_pred_train = model.predict(X_train)
Y_pred_test = model.predict(X_test)

# Printing a few predictions
print("A few Predictions: ")
for rm, lstat, medv, pred in zip(X_test["rm"], X_test["lstat"], Y_test, Y_pred_test):
    break
    
print(f"rm: {rm: .2f}, lstat: {lstat: .2f}, Actual medv: {medv:.2f}, predicted medv: {pred:.2f}")
    
```

```python
# Calculate Residuals

residual = Y_test - Y_pred_test

# Calculate and print R2 and mean squred error
r2 = r2_score(Y_test, Y_pred_test)
mse = mean_squared_error(Y_test, Y_pred_test)

print(f"R2 Score: {r2:.2f}")
print(f"Mean Squred Error: {mse:.2f}")
```

```python
# Plotting Scatter Plot for residuals
plt.figure(figsize=(10,7))

# Scatter Plot for residuals and predicted medv
sns.scatterplot(x=Y_pred_test, y=residual, color="blue")

# horizontal line to residual = 0
plt.axhline(y=0, color="red", linestyle="--")

plt.title("Rsidual Plot")
plt.xlabel("predicted medv")
plt.ylabel("Residuals")
plt.show()
```
